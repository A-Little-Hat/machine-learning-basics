# Import

```python
from sklearn.datasets import make_classification
import pandas as pd
from sklearn.preprocessing import StandardScaler
import math
from sklearn.model_selection import train_test_split
```

# Dataset

```python
x,y = make_classification(
    n_samples=50,
    n_features=2,
    n_informative=2,
    n_classes=2,
    random_state=50,
    n_repeated=0,
    n_redundant=0
)
```

convert to Dataframe

```python
x_df=pd.DataFrame(data=x)
y_df=pd.DataFrame(data=y)
```

# preprocessing

```python
x_df_scaler=StandardScaler().fit_transform(x_df)
x_df_scaler_df=pd.DataFrame(data=x_df_scaler)
x_df_scaler_df.sample(5)
```

# plot the data using matplotlib

c=y_df

```python
import matplotlib.pyplot as plt
plt.scatter(x_df_scaler_df.iloc[:,0],x_df_scaler_df.iloc[:,1],c=y_df,marker =".")
```

# train test split

```python

x_train,x_test,y_train,y_test = train_test_split(x_df_scaler_df,y_df,test_size=0.3,random_state=42)
y_train=list(y_train[0])
```

```python
y_train
```

# k value input

```python
k=int(input("enter the value of k : "))
```

## distance calculation function

```python
calcuEuclideanDistance = lambda x1,x2,y1,y2: math.sqrt(math.pow((x1-x2),2)+math.pow((y1-y2),2))
```

```python
x_df_scaler_df.sample(5)
```

# distance calculaion

```python
data=[]
for i in range(0,len(x_test)):
  dist=[]
  for j in range(0,len(x_train)):
    dist.append(calcuEuclideanDistance(x_train.iloc[j,0],x_test.iloc[i,0],x_train.iloc[j,1],x_test.iloc[i,1]))
  data.append(dist)
  
```

```python
data
```

```python
data_df=pd.DataFrame(data=data)
```

```python
data_df
```

# data process 

```python
akcsit=[]
for i in range(0,len(x_test)):
  aa=[]
  for j in range(0,len(x_train)):
    aa.append((data_df.iloc[i,j], y_train[j]))
  aa.sort(key = lambda x:x[0])
  akcsit.append(aa)
```

```python
akcsit=pd.DataFrame(data=akcsit)
```

## columns index is for train_data, rows index is for test_date

```python
akcsit
```

# no to neighbours filtering

```python
neighbours=akcsit.iloc[:,:k]
neighbours
```

# Counting the max 

```python
assignedClass=[]
for i in range(len(neighbours)):
    zeros=0
    ones=0
    for j in range(3):
        if neighbours.iloc[i][j][1]==0:
            zeros+=1
        else:
            ones+=1
    if zeros>ones:
        assignedClass.append(0)
    else:
        assignedClass.append(1)
```
