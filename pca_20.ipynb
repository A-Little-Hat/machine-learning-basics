# Program 1: 
Implement PCA with two components using scikit-learn upon the Iris data set. Fit and transform the PCA model upon the entire data set and plot the transformed data using scatter plot.

```python
import numpy as np
from sklearn import decomposition
```

```python
# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Normalize the data
from sklearn.preprocessing import normalize
X = normalize(X)
```

```python
# Create a PCA object
pca = decomposition.PCA()

# Fit the PCA object to the data
pca.fit(X)

# Get the principal components
principal_components = pca.components_

# Project the data onto the principal components
X_reduced = pca.transform(X)

# Print the principal components
print(principal_components)
```

```python
# Plot the data
import matplotlib.pyplot as plt
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

# Program 3:
Implement
PCA using scikit-learn upon the Iris data set capturing 80% variance. Fit the PCA model upon the training set and transform the model on the test test. In case the no. of components is less than 2, increase the variance captured to accommodate two components. Plot the transformed test data using scatter plot.

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

```

```python
# Load the Iris dataset
iris = load_iris()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)

```

```python
# Create a PCA model with 80% variance
pca = PCA(n_components=0.8)

# Fit the PCA model on the training set
pca.fit(X_train)

# Transform the test set using the PCA model
X_test_transformed = pca.transform(X_test)

# Check the number of components
n_components = pca.n_components_

```

```python
# If the number of components is less than 2, increase the variance captured to accommodate two components
if n_components < 2:
    pca = PCA(n_components=2)
    pca.fit(X_train)
    X_test_transformed = pca.transform(X_test)

```

```python
# Plot the transformed test data using scatter plot
plt.scatter(X_test_transformed[:, 0], X_test_transformed[:, 1], c=y_test, s=100, alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
